```python
# Note to self, had to install most recent version of seaborn to get 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
adidas_report = pd.read_excel('Adidas US Sales Datasets.xlsx', skiprows=4, index_col=1)
```

```python
adidas_report.head()
```

```python
adidas_report = adidas_report.drop(columns=['Unnamed: 0'])
adidas_report.head()
```

### Show the best selling products compared by price point to help the company determine how to set their prices going forward

```python
scatter_plot = sns.relplot(data=adidas_report, x='Price per Unit', y='Total Sales', hue='Product')
scatter_plot.fig.suptitle('Products Total Sales per Unit Price')

displot = sns.displot(adidas_report, x='Price per Unit', hue='Product', multiple='stack')
displot.fig.suptitle('Products Total Sales per Unit Price')
```

### Show the top selling retailers, by region, to help the company emphasize what brands are excelling in sales and which areas could use some attention

```python
sns.barplot(data=adidas_report, x='Retailer', y='Total Sales', hue='Region')
```

### Bring a sense of how customers are buying with the breakdown of Sales Method

```python
adidas_report.groupby(['Sales Method'])['Total Sales'].mean()
```

```python
sns.boxplot(data=adidas_report, y='Units Sold', x='Sales Method')
```

### Show the top 10 cities for Operational Profit

```python
adidas_report.groupby('City')['Operating Profit'].mean().sort_values(ascending=False)[:10]
```

### Show the bottom 5 cities for Operational Profit

```python
adidas_report.groupby('City')['Operating Profit'].mean().sort_values(ascending=False)[-5:]
```
