```python
flavors = pd.read_csv('Monthly Ice Cream Shop Sales by Flavor.csv')
flavors.head()
```

```python
# Create a line graph that displays the total units across all months (time on x axis). 
# Make sure you have the appropriate labels for the x and y axis. 
fig, ax = plt.subplots(figsize=(10, 6))
x_values = flavors['Month']
y_values = flavors['Total Units']
ax.set_xlabel('Month')
ax.set_ylabel('Total Units')
ax.set_title('Total Units Across All Months')
ax.plot(x_values, y_values)
```

```python
# Optional: Create another graph with a separate line for each ice cream flavor to compare sales by flavors over time.
fig, ax = plt.subplots(figsize=(10, 6))
x_values = flavors['Month']
y_values = flavors['Total Units']

ax.plot(x_values, flavors['Vanilla'], label='Vanilla')
ax.plot(x_values, flavors['Chocolate'], label='Chocolate')
ax.plot(x_values, flavors['Strawberry'], label='Strawberry')
ax.plot(x_values, flavors['Neapolitan'], label='Neapolitan')
ax.plot(x_values, flavors['Mango'], label='Mango')
ax.plot(x_values, flavors['Lime'], label='Lime')

ax.set_xlabel('Month')
ax.set_ylabel('Total Units')
ax.set_title('Total Units Across All Months')
ax.legend()
```

```python
# Create a bar chart that displays the total profit per month. Make sure you have the appropriate labels for the x and y axis.
fig, ax = plt.subplots(figsize=(10, 6))

ax.set_title('Total Profit Across All Months')
sns.barplot(y='Total Profit',x='Month', data=flavors)
```
